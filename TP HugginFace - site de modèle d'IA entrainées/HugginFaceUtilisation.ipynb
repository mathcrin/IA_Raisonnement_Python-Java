```python
import pandas as pd
from transformers import pipeline

# Load the models
model1 = pipeline("text-classification", model="lxyuan/distilbert-base-multilingual-cased-sentiments-student")
model2 = pipeline("text-classification", model="ac0hik/emotions_detection_french")
model3 = pipeline("text-classification", model="nlptown/bert-base-multilingual-uncased-sentiment")

# Use the models to predict the sentiment of the reviews
for i in range(min(100, len(reviews))):
    text = reviews[i]
    print(f"Review {i+1}: {text}")
    print(f"Model 1 prediction: {model1(text)}")
    print(f"Model 2 prediction: {model2(text)}")
    print(f"Model 3 prediction: {model3(text)}")
    print("\n")
```

```python
# 4 question answering
from transformers import pipeline

# Load the models
model1 = pipeline('question-answering',model="deepset/roberta-base-squad2")
model2 = pipeline('question-answering',model="distilbert/distilbert-base-uncased-distilled-squad")
model3 = pipeline('question-answering',model="Intel/dynamic_tinybert")

# Use the models to answer your questions
context="Java is a high-level, class-based, object-oriented programming language that is designed to have as few implementation dependencies as possible. It is a general-purpose programming language intended to let application developers write once, run anywhere (WORA), meaning that compiled Java code can run on all platforms that support Java without the need for recompilation."

questions = ["What is Java?", "What is the design of Java?", "What is the purpose of Java?"]
for question in questions:
    print(f"Model 1 answer: {model1(question=question, context=context)}")
    print(f"Model 2 answer: {model2(question=question, context=context)}")
    print(f"Model 3 answer: {model3(question=question, context=context)}")
```
